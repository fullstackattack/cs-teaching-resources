```python
import pandas as pd


rentals = pd.DataFrame({
    "Store": ['Downtown', 'Downtown', 'Uptown', 'Uptown', 'Suburb', 'Suburb', 'Downtown', 'Uptown'],
    "Genre": ['Action', 'Comedy', 'Action', 'Romance', 'Comedy', 'Action', 'Romance', 'Comedy'],
    "Month": ['January', 'January', 'February', 'February', 'January', 'February', 'February', 'January'],
    "Rentals": [120, 90, 150, 100, 70, 130, 80, 60],
    "Revenue": [2400, 1800, 3000, 2000, 1400, 2600, 1600, 1200]
})

# 10. Combine multiple lambdas: one for mean and one for range
combined_stats = rentals.groupby("Store")["Revenue"].agg(
    mean_revenue=lambda x: x.mean(),
    range_revenue=lambda x: x.max() - x.min()
)

combined_stats
```
